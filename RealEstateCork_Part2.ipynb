##Using Geopandas again to work on more details of addresses

```python
def find_address(df):
    # Create a GeoDataFrame from the input DataFrame
    gdf = gpd.GeoDataFrame(df, geometry='geometry', crs='EPSG:4326')
    
    # Convert the geometry to latitude and longitude
    gdf['lon'] = gdf.geometry.apply(lambda point: point.x)
    gdf['lat'] = gdf.geometry.apply(lambda point: point.y)
    
    # Make a request to the Nominatim API to find the address
    geolocator = Nominatim(user_agent='my-application')
    gdf['found_address'] = gdf.apply(lambda row: geolocator.reverse(f"{row['lat']}, {row['lon']}").address, axis=1)
    
    return gdf
```

```python
# Create a sample DataFrame
data = pd.read_csv("geocoded_try.csv")
df = pd.DataFrame(data)

# Convert the 'geometry' column to Shapely geometry object
df['geometry'] = df['geometry'].apply(lambda x: loads(x))
 
print(df)
```

##We will be using Geoseries POINTS() to collect google street view link using Latitude and Longitude

```python
def generate_street_view_links(geoseries):
    # Filter out None values
    geoseries = geoseries.dropna()
    
    # Convert point data to Shapely Point objects
    geoms = geoseries.apply(lambda x: Point(float(re.findall(r'[-+]?\d*\.\d+|\d+', str(x))[0]), float(re.findall(r'[-+]?\d*\.\d+|\d+', str(x))[1])))
    
    # Convert Point objects to GeoDataFrame
    gdf = gpd.GeoDataFrame(geometry=geoms)
    
    # Extract latitude and longitude from geometry column
    gdf["lon"] = gdf.geometry.x
    gdf["lat"] = gdf.geometry.y
    
    # Generate street view links for each point
    links = []
    for i, row in gdf.iterrows():
        lat, lon = row["lat"], row["lon"]
        link = f"https://www.google.com/maps/@{lat},{lon},3a,75y,254.1h,90t/data=!3m4!1e1!3m2!1sAF1QipOz-BpF13JhlgTmgHuyBHQE7VnJv6uNE7-UUfL6!2e10"
        links.append(link)
    
    return links
```
